```python
# Python
import pandas as pd
import numpy as np
from fbprophet import Prophet
for i in range(1,4):    
    df = pd.read_excel('hair_dryer_3.xlsx',header=0,sheet_name='Sheet'+str(i))
    # df['y'] = np.log(df['y'])
    m = Prophet()
    # m = Prophet(growth='logistic')
    m.fit(df)
    # Python
    future = m.make_future_dataframe(periods=365)
    # Python
    forecast = m.predict(future)
    forecast.to_excel('hair_dryer'+str(i)+'.xlsx')
    m.plot(forecast)
    m.plot_components(forecast).show()
```
